Q2In the above question, now scrape the following details of each product listed in first 3 pages
of your search results and save it in a dataframe and csv. In case if any product vertical has
less than 3 pages in search results then scrape all the products available under that product
vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of
Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details"
and “Product URL”. In case, if any of the details are missing for any of the product then
replace it by “-“

```python
driver=webdriver.Chrome(r"C:\Users\Admin\Desktop\chromedriver.exe")
```

```python
driver.get('http://www.amazon.in')
```

```python
#locating the search bar
search_bar=driver.find_element_by_id("twotabsearchtextbox")
search_bar.clear()
search_bar.send_keys("Mobile")
search_btn=driver.find_element_by_xpath("//span[@id='nav-search-submit-text']")
search_btn.click()
```

```python
Brand=[]
Price=[]
Exchange=[]
```

```python
Brand_list=driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
for i in Brand_list:
    Brand.append(i.text)
Brand
```

```python
Price_list=driver.find_elements_by_xpath("//span[@class='a-price-whole']")
for i in Price_list:
    Price.append(i.text)
Price
```

```python
urls=driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
ur=[]
for i in urls:
    ur.append(i.get_attribute("href"))
for url in ur:
    driver.get(url)
    try:
        Exchange_list=driver.find_elements_by_xpath("//div[@class='a-section a-spacing-none sopp-offer-container']")
        for j in Exchange_list:
            Exchange.append(j.text)
            
            
    except NoSuchElementException as e:
        Exchange.append("NO Exchange")
            
            
```

```python
Exchange
```

```python
Mb=pd.DataFrame({})
Mb["BRAND"]=Brand[:10]
Mb["PRICE"]=Price[:10]
Mb["Offer"]=Exchange[:10]
```

```python
Mb
```

```python
Mb.to_csv("Mobile_list.csv",index=False)
```
