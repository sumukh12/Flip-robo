```python
#Importing required libraries
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://bookpage.com/reviews")
print(page)
```

```python
#webpage sorce code
page.content
```

```python
#html parser
soup=BeautifulSoup(page.content,"html.parser")
print(soup)
```

```python
Book_title=soup.find_all("h4",class_="italic")
print(Book_title)
```

```python
Books=[]
for i in Book_title:
    Books.append(i.text.replace('\n',''))
print(Books)
```

```python
Author_name=soup.find_all("p",class_='sans bold')
Author_name
```

```python
Author=[]
for i in Author_name:
    Author.append(i.text.replace('\n',''))
Author
```

```python
Genre_name=[]
for i in Genre:
    Genre_name.append(i.text.replace('\n',''))
print(Genre_name)
```
