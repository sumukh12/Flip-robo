# Now i will treat the outliers by using IQR(Interquantile range)method.

```python
#for lot area
Q1=train["LotArea"].quantile(0.5)
Q3=train["LotArea"].quantile(0.95)
IQR=Q3-Q1
new_train=train[(train["LotArea"] >=Q1 -1.5*IQR) & (train["LotArea"] <=Q3 +1.5*IQR)]

#for "MasVnrArea"
Q1=train["MasVnrArea"].quantile(0.5)
Q3=train["MasVnrArea"].quantile(0.95)
IQR=Q3-Q1
new_train=train[(train["MasVnrArea"]>=Q1 - 1.5*IQR) & (train["MasVnrArea"] <=Q3 + 1.5*IQR)]

#For "BsmtFinSF1"
Q1=train["BsmtFinSF1"].quantile(0.5)
Q3=train["BsmtFinSF1"].quantile(0.95)
IQR=Q3-Q1
new_train=train[(train["BsmtFinSF1"] >=Q1 - 1.5*IQR) & (train["BsmtFinSF1"] <Q3 + 1.5*IQR)]

#For "WoodDeckSF"
Q1=train["WoodDeckSF"].quantile(0.5)
Q3=train["WoodDeckSF"].quantile(0.95)
IQR=Q3-Q1
new_train=train[(train["WoodDeckSF"] >=Q1 - 1.5*IQR) & (train["WoodDeckSF"] <Q3 + 1.5*IQR)]

#FOR "TotalBsmtSF"
Q1 = train['TotalBsmtSF'].quantile(0.5)
Q3 = train['TotalBsmtSF'].quantile(0.95)
IQR = Q3 - Q1
new_train=train[(train['TotalBsmtSF'] >= Q1 - 1.5*IQR) & (train['TotalBsmtSF'] <= Q3 + 1.5*IQR)]

#For "OpenPorchSF"
Q1 = train['OpenPorchSF'].quantile(0.5)
Q3 = train['OpenPorchSF'].quantile(0.95)
IQR = Q3 - Q1
new_train = train[(train['OpenPorchSF'] >= Q1 - 1.5*IQR) & (train['OpenPorchSF'] <= Q3 + 1.5*IQR)]

#For GarageArea
Q1=train["GarageArea"].quantile(0.5)
Q3=train["GarageArea"].quantile(0.95)
new_train=train[(train['GarageArea'] >= Q1 - 1.5*IQR) & (train['GarageArea'] <= Q3 + 1.5*IQR)]
```
